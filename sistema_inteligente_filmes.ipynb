3. ETAPA ETL: LIMPEZA

```python
print("\n" + "="*50)
print(" ETAPA ETL: LIMPANDO E ORGANIZANDO DADOS")
print("="*50)

# Fun√ß√£o para extrair o ano do t√≠tulo

def extrair_ano(titulo):
    import re
    # Procura (1995) no t√≠tulo
    ano = re.findall(r'\((\d{4})\)', titulo)
    if ano:
        return int(ano[-1])  # Pega o √∫ltimo ano encontrado
    return None

    # Fun√ß√£o para separar g√™neros
def separar_generos(generos):
    if isinstance(generos, str):
        return generos.split('|')
    return []

# Aplicar transforma√ß√µes
print(" Extraindo anos dos filmes...")
df_filmes['year'] = df_filmes['title'].apply(extrair_ano)

print(" Separando g√™neros...")
df_filmes['genres_list'] = df_filmes['genres'].apply(separar_generos)

print(" Limpando t√≠tulos...")
df_filmes['clean_title'] = df_filmes['title'].str.replace(r'\s*\(\d{4}\)', '', regex=True)

print("\n DADOS ORGANIZADOS:")
print(df_filmes[['clean_title', 'year', 'genres_list']].head())
```

üìà 4. AN√ÅLISE ESTAT√çSTICA

```python
print("\n" + "="*50)
print("üìä ETAPA DE AN√ÅLISE: ESTAT√çSTICAS E INSIGHTS")
print("="*50)

# Estat√≠sticas b√°sicas
print("\nüìä ESTAT√çSTICAS GERAIS:")
print(f"‚Ä¢ Total de filmes: {len(df_filmes)}")
print(f"‚Ä¢ Per√≠odo: {df_filmes['year'].min()} - {df_filmes['year'].max()}")
print(f"‚Ä¢ Filmes mais comuns em: {df_filmes['year'].mode().iloc[0]}")

# Contar g√™neros
print("\n DISTRIBUI√á√ÉO DE G√äNEROS:")

# Criar lista de todos os g√™neros
todos_generos = []
for generos in df_filmes['genres_list']:
    todos_generos.extend(generos)

# Contar frequ√™ncia
from collections import Counter
contagem_generos = Counter(todos_generos)

for genero, count in contagem_generos.most_common(10):
    porcentagem = (count / len(df_filmes)) * 100
    print(f"  {genero}: {count} filmes ({porcentagem:.1f}%)")
```

üß† 5. INTELIG√äNCIA ARTIFICIAL

```python
print("\n" + "="*50)
print("ü§ñ ETAPA IA: AN√ÅLISE INTELIGENTE")
print("="*50)

print("üöÄ Treinando sistema de recomenda√ß√£o b√°sico...")

# Sistema de recomenda√ß√£o SIMPLES (baseado em g√™neros)
class RecomendadorFilmes:
    def __init__(self, dados_filmes):
        self.dados = dados_filmes.copy()

    def recomendar_por_genero(self, generos_favoritos, n=5):
        """Recomenda filmes baseado em g√™neros"""
        resultados = []

        for idx, filme in self.dados.iterrows():
            generos_filme = set(filme['genres_list'])
            generos_user = set(generos_favoritos)

            # Calcular similaridade
            similaridade = len(generos_filme.intersection(generos_user))

            if similaridade > 0:
                resultados.append({
                    'filme': filme['clean_title'],
                    'ano': filme['year'],
                    'generos': filme['genres'],
                    'score': similaridade
                })

        # Ordenar por score
        resultados.sort(key=lambda x: x['score'], reverse=True)
        return resultados[:n]

    def analisar_tendencias(self):
        """Analisa tend√™ncias por ano"""
        tendencias = {}
        for ano in df_filmes['year'].unique():
            if pd.notna(ano):
                filmes_ano = df_filmes[df_filmes['year'] == ano]
                generos_ano = []
                for g in filmes_ano['genres_list']:
                    generos_ano.extend(g)

                top_generos = Counter(generos_ano).most_common(3)
                tendencias[ano] = top_generos

        return tendencias

# Criar recomendador
recomendador = RecomendadorFilmes(df_filmes)

# Testar recomenda√ß√µes
print("\nüéØ RECOMENDA√á√ïES PERSONALIZADAS:")
print("Para quem gosta de: ['Comedy', 'Romance', 'Drama']")

recomendacoes = recomendador.recomendar_por_genero(['Comedy', 'Romance', 'Drama'], n=3)

for i, rec in enumerate(recomendacoes, 1):
    print(f"\n{i}. {rec['filme']} ({rec['ano']})")
    print(f"   G√™neros: {rec['generos']}")
    print(f"   Compatibilidade: {'‚≠ê' * rec['score']}")

# An√°lise de tend√™ncias
print("\nüìà TEND√äNCIAS POR ANO:")
tendencias = recomendador.analisar_tendencias()
for ano, top_generos in tendencias.items():
    if top_generos:
        generos_str = ", ".join([f"{g[0]}({g[1]})" for g in top_generos])
        print(f"{ano}: {generos_str}")
```
